1. Llamamos a las librerias que necesitamos para utilizar K-Mean

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler
from sklearn import metrics
from sklearn.metrics import pairwise_distances
from sklearn import datasets
import random
plt.figure(figsize=(12, 12))
```

Creamos un dataframe llamado wine a partir de nuestro csv y mostramos la cabecera (variables de nuestro dataframe).

```
wine=pd.read_csv('wine.csv')
```

```
wine.info()
```

```
wine.head
```

Se procede a eliminar la variable class, ya que no puede ser utilizada según las instrucciones del ejercicio. Nuestro dataframe, tendrá ahora 178 observaciones y 13 columnas.

```
wine.drop(['Class'], axis=1)
```

Se decide hacer una clusterización en función de las variables alcohol e intensidad del color a través de Kmeans++. 
Para ello se crea un nuevo dataframe con ambas variables, para no crear perjuicio en el archivo original. Se procede a eliminar los posibles valores perdidos que existieran en las observaciones. Además las ordeno de menor a mayor.
 

```
wine_dataw = wine[['Alcohol','Color intensity']].copy(deep=True)
wine_dataw.dropna(axis=0, inplace=True)
wine_dataw.sort_values(by=['Alcohol','Color intensity'], inplace=True)
cluster_array = np.array(wine_dataw)
```

Se puede mostrar el cabecero de nuestro dataframe, donde aparece el vino con menor alcohol e intensidad en el color (un blanco afrutado posiblemente) porque se lo dijimos anteriormente que las ordenara. Y dibujar la distribución de las observaciones (eje x alcohol; eje y intensidad) a partir del array bidimensional creado anteriormente.

```
print(wine_dataw.head())
plt.scatter(cluster_array[:,0],cluster_array[:,1])
plt.title("Análisis previo Intensidad Color vs Alcohol")
```

El método kmeans++ que se va a utilizar mejora la elección de la semilla inicial, ya que esta es muy relevante para el resultado final. Kmeans++ busca de forma más o menos eficiente los centroides, en función de la distribución de datos. Otra ventaja es la reducción en el número de iteracciones.
Se decide realizar una agrupación de k=2, y dibujarlo en un plano geom point para ver la clusterización y distribución de las observaciones.
Vemos que existen observaciones menos observaciones por encima del valor seis dentro de la categorización de la intensidad de color y la gran mayoría por debajo de esta. Y la graduación alcohólica situa a la mayoría de las observaciones entre 12 y 14.

```
KM=KMeans(n_clusters=2)
y_pred=KM.fit_predict(cluster_array)
plt.scatter(cluster_array[:,0],cluster_array[:,1],c=y_pred)
plt.title("Clúster KMeans++")
print(KM.cluster_centers_)
```

Las variables no tienen valores muy elevados, pero sería posible normalzar los datos para eliminar el efecto escala. Es por eso que se decide aplicar la estandarización para normalizar (aunque existen otros métodos como el máximo o el mínimo-máximo).

```
standar = StandardScaler()
standar.fit(cluster_array)
#Escalamos los datos
scaled_data = standar.transform(cluster_array)
plt.scatter(scaled_data[:,0],scaled_data[:,1])
plt.title("Datos Estandarizados")
```

Si calculásemos su media sería cero y la desviación típica 1.

Se aplica el mismo método de clusterización de KMeans++, y se grafica.


```
KM=KMeans(n_clusters=2)
y_pred=KM.fit_predict(scaled_data)
plt.scatter(scaled_data[:,0],scaled_data[:,1],c=y_pred)
plt.title("Clúster KMeans++ Estandarizado")
print(KM.cluster_centers_)
```

Se muestra que la primera clusterización antes de tipificar, se agrupaban más elementos entorno al eje x que es el alcohol. Y ahora se compensa un poco más (aunque las observaciones se posicionán muy parecidas).

Mediante el siguiente comando podemos calcular el Coeficiente e Silhouette, donde podríamos realizarlo con varios grupos y ver cual es mejor (k= 2, 3, ..)

```
labels = KM.labels_
metrics.silhouette_score(scaled_data, labels, metric='euclidean')
```
